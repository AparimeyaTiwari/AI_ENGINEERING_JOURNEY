# Neural Network Training Pipeline - 6 Stages

![Neural Network Pipeline Diagram](BASIC_LIBRARIES/pytorch_learning/Screenshot%202026-01-07%20at%2014.24.10.png)

## 1) Data Ingestion

Collecting raw info or data and organizing it.

**Example:** Delivery predictor â†’ entries from delivery records

**Problem: Data is messy (this is the norm)**
- Different data types: "22 minutes" vs 22.0
- Incomplete data
- Missing values
- Negative delivery times
- Records showing impossible speeds (200 mph on a bike)

We must clean the data in the next step.

---

## 2) Data Preparation

Clean + Organize + Transform

- Handling missing values
- Fixing other errors (removing impossible delivery times, duplicate entries)
- Engineering new features
  - Example: Converting addresses like "123 Oak Street" into distances like 8.2 miles

**Note:** Most models fail since the data was messy, not the math.

This stage takes the longest time and requires the most code.

---

## 3) Model Building

Choosing the architecture:
- Number of neurons
- Hidden layers
- How neurons are connected

**Example:**
```python
model = nn.Sequential(nn.Linear(1,1))
```

**Good thing:** Easy to do even for complex data - just one line.

---

## 4) Training

Passing in inputs and getting outputs.

Here our model starts understanding underlying patterns and trends.

**Example inputs:**
- 8.2 miles â†’ 22 minutes
- 12.5 miles â†’ 31 minutes

The model gradually figures out the pattern.

**Main part where we can say the model is identifying inner details.**

---

## 5) Evaluation

How the model performs in real cases.

Use a set known as **testing set** (left out initially during training).

**Example:**
- Model predicts: 28 minutes for 15 miles
- Reality: 32 minutes

Questions:
- How often is the model close?
- How often is it way off?
- Does it work well enough to trust it?

---

## 6) Deployment

Getting model out to real-world users.

Last stage - convert this to production use.


````python
import torch
import torch.nn as nn
import torch.optim as optim

import helper_utils


# This line ensures that your results are reproducible and consistent every time.
torch.manual_seed(42)

'''
Pytorch core functions -> torch
Component to build neural network -> nn
Tools for training the network -> optim
'''

distances = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)
times = torch.tensor([[6.96], [12.11], [16.77], [22.21]], dtype=torch.float32)

'''Here
[[6.96],[12.27],[16.77],[22.21]] -> subset of entire dataset, collection of data point => BATCH
[6.96] => Sample in the batch. Since we are only considering distance 1 feature in the bracket
dtype = torch.float32 => Tells torch what type of numbers to store
'''

model = nn.Sequential(nn.Linear(1,1))

'''(1,1)=> first 1: the input 
        second 1: the output (delivery times)
'''

loss_function = nn.MSELoss() # to calculate the loss ie how wrong or right the prediction our, have to reduce this
optimizer = optim.SGD(model.parameters(),lr = 0.01) 
# Stocastic gradeint descent, algo which decides which direction to adjust weight and bias to reduce error or loss
# model paramter: to access those values ie weight and bias
# lr => learning rate (how we are adjusting: too much or small increments)

#training loop (automating manual changes in weights and biases we did)
# each run through the entire training data=> EPOCH
# In a EPOCH
'''
1) make prediction
2) Measure how off the prediction is
3) adjust the internal parameters (ex) weight and bias
'''
for epoch in range(500):
    #reseting optimizer (so that we start fresher for each epoch ie we do not want to keep old adjustments from teh previosu steps)
    optimizer.zero_grad()

    # 1) Make prediction
    output = model(distances)

    # 2)Calculate the loss
    loss = loss_function(output,times)

    # 3) Calculate adjustments => Backward prop
    loss.backward() 

    #4) Update the model => Makes the adjustments
    optimizer.step()

    if((epoch+1)%25 == 0):
        print(f"The loss for epoch{epoch+1} is {loss.item()}")

helper_utils.plot_results(model, distances, times)

with torch.no_grad(): #ie only doing infernece for now
    test_input = torch.tensor([[15.0]],dtype=torch.float32)
    predicted_time = model(test_input)
    print(f"Predicted time is: {predicted_time.item():.1f} minutes")

    if predicted_time.item() > 30:
        print("\nDecision: Do NOT take the job. You will likely be late.")
    else:
        print("\nDecision: Take the job. You can make it!")


layer = model[0]

weight = layer.weight.data.numpy()
bias  = layer.bias.data.numpy()

print(weight)
print(bias)
````

## Why Linear Models Break Down

From the earlier diagram, we can see that **real-world relationships are not linear**.

Examples:
- Traffic might increase **after 2 miles**
- You might **enter the city**, changing conditions suddenly

Such patterns **cannot be captured by a straight line**.

---

## First Idea: Add More Neurons

To learn more complex patterns, we add **more neurons**.

Result:
- Neurons are grouped into an intermediate layer
- This intermediate layer is called a **Hidden Layer**

At this point, the network *appears* more powerful.

---

## The Core Issue

Even with:
- Multiple neurons  
- Multiple hidden layers  

The model is still performing only **linear transformations**:

\[
y = Wx + b
\]

### Important Insight
A composition of linear functions is **still linear**.

> **Linear transformations cannot represent non-linear patterns.**

---

## Solution: Activation Functions

To overcome this limitation, we introduce **Activation Functions**.

### What an Activation Function Does
1. Applied **after** the linear transformation  
2. Introduces **non-linearity** into the network  
3. Enables learning of **complex curves and patterns**

Without activation functions:
- Neural networks reduce to linear models
- Depth becomes meaningless

---

## ReLU (Rectified Linear Unit)

For now, we focus on the most widely used activation function: **ReLU**.

### Definition

\[
\text{ReLU}(x) =
\begin{cases}
0 & \text{if } x \le 0 \\
x & \text{if } x > 0
\end{cases}
\]

### Intuition
- Negative inputs are **suppressed**
- Positive inputs **flow unchanged**
- This introduces bends into the model, enabling **non-linear learning**

````python
import torch
import torch.nn as nn 
import torch.optim as optim

from helper_utils import *


# Combined dataset: bikes for short distances, cars for longer ones
distances = torch.tensor([
    [1.0], [1.5], [2.0], [2.5], [3.0], [3.5], [4.0], [4.5], [5.0], [5.5],
    [6.0], [6.5], [7.0], [7.5], [8.0], [8.5], [9.0], [9.5], [10.0], [10.5],
    [11.0], [11.5], [12.0], [12.5], [13.0], [13.5], [14.0], [14.5], [15.0], [15.5],
    [16.0], [16.5], [17.0], [17.5], [18.0], [18.5], [19.0], [19.5], [20.0]
], dtype=torch.float32)

# Corresponding delivery times in minutes
times = torch.tensor([
    [6.96], [9.67], [12.11], [14.56], [16.77], [21.7], [26.52], [32.47], [37.15], [42.35],
    [46.1], [52.98], [57.76], [61.29], [66.15], [67.63], [69.45], [71.57], [72.8], [73.88],
    [76.34], [76.38], [78.34], [80.07], [81.86], [84.45], [83.98], [86.55], [88.33], [86.83],
    [89.24], [88.11], [88.16], [91.77], [92.27], [92.13], [90.73], [90.39], [92.98]
], dtype=torch.float32)
````

````python
#NORMALIZING DATA

#normalizing distances
distances_mean = distances.mean()
distances_std = distances.std()

#normalizing times
times_mean = times.mean()
times_std = times.std()

distances_norm = (distances-distances_mean)/distances_std
times_norm = (times-times_mean)/times_std
````

````python
helper_utils.plot_data(distances_norm, times_norm, normalize=True)
````

````python
torch.manual_seed(27)
model = torch.nn.Sequential(
    nn.Linear(1,3),
    nn.ReLU(),
    nn.Linear(3,1)
)

loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(),lr=0.01)

for epoch in range(1000):

    optimizer.zero_grad()

    pred = model(distances_norm)

    loss = loss_function(pred,times_norm)

    loss.backward()

    optimizer.step()

    if (epoch + 1) % 50 == 0:
        helper_utils.plot_training_progress(
            epoch=epoch,
            loss=loss,
            model=model,
            distances_norm=distances_norm,
            times_norm=times_norm
        )

print("\nTraining Complete.")
print(f"\nFinal Loss: {loss.item()}")
````

````python
helper_utils.plot_final_fit(model, distances, times, distances_norm, times_std, times_mean)
````

````python
dist_to_pred = 8.1

with torch.no_grad():
    distance_input = torch.tensor([[dist_to_pred]],dtype=torch.float32)
    distance_input_norm = (distance_input-distances_mean)/distances_std

    times_output_norm = model(distance_input_norm)
    times_output_actual = (times_output_norm*times_std)+times_mean

    print(f"Prediction for a {dist_to_pred}-mile delivery: {times_output_actual.item():.1f} minutes")
    
    # First, check if the delivery is possible within the 45-minute timeframe
    if times_output_actual.item() > 45:
        print("\nDecision: Do NOT promise the delivery in under 45 minutes.")
    else:
        # If it is possible, then determine the vehicle based on the distance
        if dist_to_pred <= 3:
            print(f"\nDecision: Yes, delivery is possible. Since the distance is {dist_to_pred} miles (<= 3 miles), use a bike.")
        else:
            print(f"\nDecision: Yes, delivery is possible. Since the distance is {dist_to_pred} miles (> 3 miles), use a car.")
````
